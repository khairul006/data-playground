```python
import json

log_file_path = os.path.join('..', 'data', 'app.log')
extracted_data = []

with open(log_file_path, 'r') as file:
    for line in file:
        # 1. Find where the JSON starts
        # Your log format uses 'Consumed json message: ' as the delimiter
        if "Consumed json message: " in line:
            parts = line.split("Consumed json message: ")
            
            if len(parts) > 1:
                json_string = parts[1].strip()
                
                try:
                    # 2. Convert string to Python Dictionary
                    data_dict = json.loads(json_string)
                    extracted_data.append(data_dict)
                except json.JSONDecodeError:
                    print(f"Failed to parse line: {line}")

# View the first extracted object
if extracted_data:
    print(f"Successfully extracted {len(extracted_data)} records.")
    print(f"Sample: {extracted_data[0]}")
```

```python
print(type(extracted_data))
# Output: <class 'list'>

first_item = extracted_data[0]
print(type(first_item))

print(type(first_item['info']))
# Output: <class 'dict'>

print(type(first_item['siteId']))
# Output: <class 'str'> (or 'int' depending on the JSON)
```
